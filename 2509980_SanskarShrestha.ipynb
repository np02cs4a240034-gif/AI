```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("World Happiness Report.csv")
df.head()
```

```python
# 1. Load first 10 rows
df.head(10)

# 2. Number of rows & columns
df.shape

# 3. Columns & data types
df.dtypes
```

```python
# Mean, Median, SD of Score
mean_score = df["score"].mean()
median_score = df["score"].median()
std_score = df["score"].std()

mean_score, median_score, std_score

# Highest & Lowest scoring countries
highest = df.loc[df["score"].idxmax()]
lowest = df.loc[df["score"].idxmin()]

highest, lowest
```

```python
df.isna().sum()
```

```python
filtered = df[df["score"] > 7.5]
top10_sorted = filtered.sort_values("Log GDP per capita", ascending=False).head(10)
top10_sorted
```

```python
def category(score):
    if score < 4:
        return "Low"
    elif score <= 6:
        return "Medium"
    else:
        return "High"

df["Happiness Category"] = df["score"].apply(category)
df.head()
```

```python
top10 = df.sort_values("score", ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x="score", y="Country name", data=top10)
plt.title("Top 10 Happiest Countries")
plt.show()
```

The gaps among the best-ranked countries are often very little and this may probably indicate that the nations with the highest scores generally possess good well-being across various metrics.

```python
south_asia = ["Afghanistan","Bangladesh","Bhutan","India",
              "Maldives","Nepal","Pakistan","Sri Lanka"]

south_df = df[df["Country name"].isin(south_asia)].copy()
south_df.to_csv("SouthAsia.csv", index=False)
south_df
```

```python
south_df["Composite Score"] = (
    0.40 * south_df["Log GDP per capita"] +
    0.30 * south_df["Social support"] +
    0.30 * south_df["Healthy life expectancy"]
)

ranked = south_df.sort_values("Composite Score", ascending=False)
ranked
```

```python
top5 = ranked.head(5)
plt.figure(figsize=(8,5))
sns.barplot(x="Composite Score", y="Country name", data=top5)
plt.title("Top 5 South Asian Countries by Composite Score")
plt.show()
```

# Horizontal Bar Chart  South Asia Composite Score Ranking

**Interpretation**:
The composite ranking based on weighted metrics (GDP, Social Support, Healthy Life Expectancy) is illustrated in this bar chart. The top of the ranking belongs to the countries that have stronger development and better social support systems. The chart reveals the discrepancies that do not necessarily correspond with the original happiness score ranking, thus indicating that the multi-factor scoring may alter the country positions. Nations that exhibit poor health or economic performance are usually ranked lower.

```python
Q1 = south_df["Log GDP per capita"].quantile(0.25)
Q3 = south_df["Log GDP per capita"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = south_df[(south_df["Log GDP per capita"] < lower) | (south_df["Log GDP per capita"] > upper)]
outliers
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=south_df, x="Log GDP per capita", y="score", label="Normal")
sns.scatterplot(data=outliers, x="Log GDP per capita", y="score", color="red", s=100, label="Outlier")
plt.title("Outliers in South Asia")
plt.legend()
plt.show()
```

# **Scatter Plot  South Asia's Outliers in Economic Terms (GDP vs. Score**)

Thus, the countries of South Asia have the same GDP and happiness traits with no great deviations. The steady grouping indicates the presence of similarities in the regions economic and social welfare

```python
south_df["GDP-Score Gap"] = south_df["Log GDP per capita"] - south_df["score"]

largest = south_df.sort_values("GDP-Score Gap", ascending=False).head(3)
smallest = south_df.sort_values("GDP-Score Gap").head(3)

plt.figure(figsize=(8,4))
sns.barplot(x="GDP-Score Gap", y="Country name", data=largest)
plt.title("Largest Positive Gap")
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(x="GDP-Score Gap", y="Country name", data=smallest)
plt.title("Largest Negative Gap")
plt.show()
```

# **Interpretation of GDPScore Gap**

**# Largest Positive Ga**p

The nations where the largest positive gaps are registered have higher GDP per capita but lower happiness scores. It indicates that the power of the economy is not reaching people's well-being measures, perhaps owing to reasons like inequality, stress, or lack of social support.

**# Largest Negative Gap**

The nations with the largest negative gaps depict a scenario where GDP is less but happiness is more. Strong community ties, cultural values or better social environments might be the factors that enhance happiness even in the presence of limited economic resources.

```python
def find_outliers(column):
    Q1 = me_df[column].quantile(0.25)
    Q3 = me_df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = me_df[(me_df[column] < lower_bound) | (me_df[column] > upper_bound)]
    return outliers

me_out_score = find_outliers("score")
me_out_gdp = find_outliers("Log GDP per capita")

plt.figure(figsize=(8,6))
plt.scatter(me_df["Log GDP per capita"], me_df["score"], label="Normal")

plt.scatter(me_out_score["Log GDP per capita"], me_out_score["score"],
            s=120, edgecolors='black', label="Score Outlier")

plt.scatter(me_out_gdp["Log GDP per capita"], me_out_gdp["score"],
            s=120, edgecolors='black', label="GDP Outlier")

plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.title("Middle East Outliers")
plt.legend()
plt.show()
```

**The scatter plot illustrates how GDP per capita correlates with the happiness score of Middle Eastern nations**. A majority of the points are within the standard range, whereas the outlier markers are the ones highlighted, which were found using the 1.5IQR rule. The score outliers indicate those countries that have very high or very low happiness levels, and the GDP outliers show the countries whose economic values are extremely different from the regional average. The outliers play a role in revealing nations whose economic or happiness conditions are contrary to the usual Middle Eastern ones.

```python
plt.figure(figsize=(8,6))
plt.boxplot([south_df["score"], me_df["score"]],
            tick_labels=["South Asia", "Middle East"])
plt.title("Score Distribution Comparison")
plt.ylabel("Happiness Score")
plt.show()
```

# Interpretation of Boxplot

The boxplot makes a comparison between the two regions, South Asia and the Middle East, regarding the distribution of happiness scores. The Middle East, in general, presents a higher median score and a broader range, which suggests that there is a greater difference in the happiness levels across the whole region. On the contrary, South Asia has a lower median and a tighter distribution implying that the countries in the region have similar levels of happiness but the overall score is lower.
